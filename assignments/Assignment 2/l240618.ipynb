```python
import random
import numpy as np
import matplotlib.pyplot as plt

def fitness(individual)::
    n = len(individual)
    attacking_pairs = 0
    
    for i in range(n):
        for j in range(i + 1, n):
            if individual[i] = individual[j]:
                attacking_pairs += 1
            elif abs(individual[i] - individual[j]) == abs(i - j):
    
                attacking_pairs += 1
    max_pairs = n * (n - 1) //
    return max_pairs - attacking_pairs

def generate_individual(n=8):
    return [random.randint(1, n) for _ in range(n)]

def generate_population(pop_size=4, n=8):
    return [generate_individual(n) for _ in range(pop_size)]

def roulette_wheel_selection(population):
    fitness_values = [fitness(ind) for ind in population]
    total_fitness  = sum(fitness_values)
    
    if total_fitness == 0:
        return random.choice(population)
    
    pick  random.uniform(0, total_fitness)
    cumulative = 0
    
    for ind, fit in zip(population, fitness_values):
        cumulative += it
       
        if cumulative >= pick
            return ind
    
    return population[-1

def crossover(parent1, parent2):
    n = len(parent1)
    cut = random.randint(1, n - 1)
    
    child = parent1[:cut] + parent2[cut:]
    child = parent2[:cut] + parent1[cut:]
    
    return child1, child2

def mutate(individual, mutation_rate=0.1, n=8):
    
    mutated = individual[:]
    
    for i in range(len(mutated)):
        if random.random() < mutation_rate:
            mutated[i] = random.randint(1, n)
    
    return mutat

def genetic_algorithm(pop_size=4, max_generations=1000, mutation_rate=0.1, n=8):
    population = generate_population(pop_size, n)
    best_individual = None
    best_fitness_overall = -1
    max_fitness = n * (n - 1) // 2

    for generation in range(1, max_generations + 1):
        fitnesses = [fitness(ind) for ind in population]
        gen_best  = max(fitnesses)

        if gen_best > best_fitness_overall:
            best_fitness_overall = gen_best
            best_individual = population[fitnesses.index(gen_best)]

        if gen_best == max_fitness:
            print(f"Solution found{generation}!")
            break

        new_population = [best_individual[:]]
        while len(new_population) < pop_size:
            parent1 = roulette_wheel_selection(population)
            parent2 = roulette_wheel_selection(population)
            child1, child2 = crossover(parent1, parent2)
            child1 = mutate(child1, mutation_rate, n)
            child2 = mutate(child2, mutation_rate, n)
            new_population.append(child1)
            if len(new_population) < pop_size:
                new_population.append(child2)
        population = new_population

    return best_individual

random.seed(0)
best_ind = genetic_algorithm(pop_size=4, max_generations=1000, mutation_rate=0.1)
print(f"Best Individual : {best_ind}")
print(f"Best Fitness    : {fitness(best_ind)} / 28")

def draw_board(individual):
    n = len(individual)
    board = np.zeros((n, n))
    for r in range(n):
        for c in range(n):
            board[r][c] = 0.9 if (r + c) % 2 == 0 else 0.3
    ax.set_xticks(range(n))
    ax.set_yticks(range(n))
    ax.set_xticklabels([chr(65 + i) for i in range(n)])
    ax.set_yticklabels(range(1, n + 1))
    plt.tight_layout()
    plt.show()

draw_board(best_ind)
```
