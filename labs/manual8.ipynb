```python
#PART C
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import numpy as np

salaries = np.array([[25000], [42000], [60000], [85000], [120000], [200000]])

#21
scaler_mm = MinMaxScaler()
sal_norm = scaler_mm.fit_transform(salaries)
print("\nNormalized Salaries:\n", sal_norm)

#22
scaler_std = StandardScaler()
sal_std = scaler_std.fit_transform(salaries)
print("Standardized Salaries:\n", sal_std)

#23
# Normalization Formula: (X - X_min) / (X_max - X_min)
# Standardization Formula: (X - mean) / std
print("Standardized Mean:", sal_std.mean().round(10))

#24 StandardScaler is more appropriate here because salary data often contains outliers
```

```python
#PARTC
#QUESTION 7

from sklearn.model_selection import train_test_split
X = np.array([[5], [10], [15], [20], [25], [30], [35], [40], [45],[50]])

#25
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)

#26
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
print("\nCorrectly Scaled Test:\n", X_test_scaled)

#27
X_test_incorrect = scaler.fit_transform(X_test)
print("Incorrectly Scaled Test:\n", X_test_incorrect)

#28 Data leakage occurs when information from outside the training dataset is used to create the model. Calling fit_transform on test data recomputes mean/std based on the test set, leaking its distribution into the scaling process

```

```python
#PART C 
#QUESTION 8
import numpy as np
from scipy.stats import skew
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Data initialization
A = np.array([10, 12, 11, 13, 10, 12, 14, 11, 13, 12]).reshape(-1, 1)
B = np.array([1, 2, 1, 3, 2, 1, 100, 2, 1, 3]).reshape(-1, 1)

#29
skew_a = skew(A)[0]
skew_b = skew(B)[0]

print(f"Skewness of Feature A: {skew_a:.4f}")
print(f"Skewness of Feature B: {skew_b:.4f}")

#30,31
if abs(skew_a) < 0.5:
    # Justification: Since the skewness is between -0.5 and 0.5, the distribution is 
    # considered roughly symmetric. StandardScaler is preferred for symmetric data.
    scaler_a = StandardScaler()
    A_scaled = scaler_a.fit_transform(A)
    print("\nFeature A Scaled (StandardScaler):\n", A_scaled.flatten())
else:
    scaler_a = MinMaxScaler()
    A_scaled = scaler_a.fit_transform(A)
    print("\nFeature A Scaled (MinMaxScaler):\n", A_scaled.flatten())

# Feature B logic
if abs(skew_b) < 0.5:
    scaler_b = StandardScaler()
    B_scaled = scaler_b.fit_transform(B)
    print("\nFeature B Scaled (StandardScaler):\n", B_scaled.flatten())
else:
    # Justification: The skewness is significantly greater than 0.5 (heavy right skew). 
    # MinMaxScaler is used to bound skewed data within a specific range [0, 1].
    scaler_b = MinMaxScaler()
    B_scaled = scaler_b.fit_transform(B)
    print("\nFeature B Scaled (MinMaxScaler):\n", B_scaled.flatten())
```

```python
#BONUS
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder, StandardScaler, MinMaxScaler

#38 Creating initial data fram
data = {
    'CustomerID': [1, 2, 3, 4, 5, 6, 7, 8, 2, 5],
    'Age': [25, None, 34, 28, None, 45, 31, 27, None, None],
    'Income': ['50000', '75000', 'N/A', '62000', '48000', '91000', '55000', 'N/A', '75000', '48000'],
    'City': ['Lahore', 'Karachi', None, 'Islamabad', 'Lahore', 'Karachi', None, 'Lahore', 'Karachi', 'Lahore'],
    'Membership': ['Silver', 'Gold', 'Bronze', 'Gold', 'Silver', 'Platinum', 'Bronze', 'Silver', 'Gold', 'Silver'],
    'Spent': [1200, 4500, 800, 3200, 1100, 9800, 750, None, 4500, 1100]
}
df = pd.DataFrame(data)

#39 removing duplicate rows on ID and reseting the index
df.drop_duplicates(subset='CustomerID', keep='first', inplace=True)
df.reset_index(drop=True, inplace=True)

#40 Converting Income to numeric and fill missing Income with medians
df['Income'] = pd.to_numeric(df['Income'], errors='coerce')
df['Income'].fillna(df['Income'].median(), inplace=True)
df['Spent'].fillna(df['Spent'].median(), inplace=True)

#41 Filling missing Age with median and missing City with mode
df['Age'].fillna(df['Age'].median(), inplace=True)
df['City'].fillna(df['City'].mode()[0], inplace=True)

#42 Apply OrdinalEncoder to Membership
rank_order = [['Bronze', 'Silver', 'Gold', 'Platinum']]
encoder = OrdinalEncoder(categories=rank_order)
df['Membership'] = encoder.fit_transform(df[['Membership']])

#43 Using pd.get_dummies()
df = pd.get_dummies(df, columns=['City'], prefix='City')

# 44. Compute skewness and threshold is 0.5
for col in ['Income', 'Age']:
    skew_val = df[col].skew()
    print(f"Skewness of {col}: {skew_val:.2f}")
    
    if abs(skew_val) < 0.5:
        scaler = StandardScaler()
        print(f"Applying StandardScaler to {col}")
    else:
        scaler = MinMaxScaler()
        print(f"Applying MinMaxScaler to {col}")
    
    df[col] = scaler.fit_transform(df[[col]])

# 45. Print final results
print("\nFinal DataFrame Shape:", df.shape)
print("Missing values per column:\n", df.isnull().sum())
print("\nFinal Preprocessed DataFrame:\n", df.head())
```
