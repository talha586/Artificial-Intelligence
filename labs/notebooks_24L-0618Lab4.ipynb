```python
##QUESTION 1
# import numpy as np
# array=np.random.randint(1,50,(4,4))

# #a)
# print("Original Array: \n",array)
# print("Shape of Array: ",array.shape)
# print("Dimension of Array: ",array.ndim)
# print("Data type of Array: ",array.dtype)

# #b)
# sub_array=array[0:2,2:4]
# print("Sub Array: \n",sub_array)

# #c)
# square_array=np.square(array)
# print("Square of Array: \n",square_array)

# #d)
# row_means=np.mean(array,axis=1)
# print("Row Means: \n",row_means)

##QUESTION 2
# import numpy as np
# array=np.array([[10,20,30],
#                 [40,50,60],
#                 [70,80,90]])
# #a)
# sub_array=array[0:3,1:2]
# print(sub_array)

# #b)
# square_root_array=np.sqrt(array)
# print(square_root_array)

# #c)
# row_sum=np.sum(array,axis=1)
# print(row_sum)

##QUESTION 3

# import pandas as pd

# read_student=pd.read_csv('students.csv')
# read_student=read_student.sort_values(by='Marks', ascending=False)
# read_student['Grade']=read_student['Marks'].apply(lambda x: 'A' if x>85 else 'B')
# print(read_student)

# mean_val=read_student['Marks'].mean()
# median_val=read_student['Marks'].median()
# min_val=read_student['Marks'].min()

# print("Mean Value: ",mean_val)
# print("Median Value: ",median_val)
# print("Min Value: ",min_val)

#QUESTION 4
import numpy as np
import pandas as pd
read_iris=pd.read_csv("iris.csv")
print("First 10 rows: ")
print("Dataset Shape: ",read_iris.shape)
print("Datatype: ",read_iris.dtypes)
print("Data Columns: ",read_iris.columns)
print("Non Null Values: ",read_iris.count())

filtered_species=read_iris.loc[read_iris['Species']=='Setosa']
filtered_sorted_species=filtered_species.sort_values(by='PetalLength',ascending=False)
filtered_sorted_species['Petal Area']=filtered_sorted_species['PetalLength']*filtered_sorted_species['PetalWidth']
print(filtered_sorted_species.head(10))

#QUESTION 5
mean_val=read_iris['SepalLength'].mean()
median_val=read_iris['SepalLength'].median()
mode_val=read_iris['SepalLength'].mode()
range_val=read_iris['SepalLength'].max()-read_iris['SepalLength'].min()
print("Mean: ",mean_val) 
print("Median: ",median_val) 
print("Mode: ",mode_val)
print("Range:",range_val)

Q1 = read_iris['SepalLength'].quantile(0.25)
Q3 = read_iris['SepalLength'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = read_iris[(read_iris['SepalLength'] < lower_bound) or (read_iris['SepalLength'] > upper_bound)]

print("Outlier")
if outliers :
    print("No outliers")
else:
    print(outliers)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

read_iris = pd.read_csv('iris.csv')

read_iris.plot.scatter(x='SepalLength', y='SepalWidth', title='Sepal Length vs Sepal Width')

read_iris['PetalLength'].plot.hist(title='Histogram of Petal Length')
plt.xlabel('PetalLength')
plt.show()

read_iris['Species'].value_counts().sort_values().plot.bar(title='Count of Each Species')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

read_iris.select_dtypes(include='number').boxplot()
plt.title('Boxplot for Numerical Features')
plt.show()
```
