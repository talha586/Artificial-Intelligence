```python
import random

def evaluate_expression(x, y, z):
    return 6*x**3 + 9*y**2 + 90*z - 25

def fitness(x, y, z):
    ans = evaluate_expression(x, y, z)
    if ans == 0:
        return 99999
    else:
        return abs(1 / ans)

initial_population = []
for _ in range(1000):
    x, y, z = random.uniform(0, 1000), random.uniform(0, 1000), random.uniform(0, 1000)
    initial_population.append((fitness(x, y, z), (x, y, z)))

initial_population.sort(key=lambda sol: sol[0], reverse=True)

top_100 = initial_population[:100]
variables = []
for sol in top_100:
    variables.extend([sol[1][0], sol[1][1], sol[1][2]])

new_generation = []
for _ in range(1000):
    nx = random.choice(variables) * random.uniform(0.99, 1.01)
    ny = random.choice(variables) * random.uniform(0.99, 1.01)
    nz = random.choice(variables) * random.uniform(0.99, 1.01)
    new_generation.append((fitness(nx, ny, nz), (nx, ny, nz)))

new_generation.sort(key=lambda sol: sol[0], reverse=True)
print(new_generation[0])
```

```python
import random

def evaluate_expression(x, y, z):
    return 6*x**3 + 9*y**2 + 90*z - 25

def fitness(x, y, z):
    ans = evaluate_expression(x, y, z)
    if ans == 0:
        return 99999
    else:
        return abs(1 / ans)

solutions = [(random.uniform(0, 1000), random.uniform(0, 1000), random.uniform(0, 1000)) for _ in range(1000)]

for gen in range(10000):
    ranked = []
    for s in solutions:
        ranked.append((fitness(s[0], s[1], s[2]), s))
    
    ranked.sort(key=lambda x: x[0], reverse=True)
    
    print(f"Gen {gen}: Best Fitness = {ranked[0][0]}, Solution = {ranked[0][1]}")
    
    if ranked[0][0] > 999:
        break
        
    best_solutions = ranked[:100]
    pool = []
    for bs in best_solutions:
        pool.extend([bs[1][0], bs[1][1], bs[1][2]])
        
    new_gen = []
    for _ in range(1000):
        nx = random.choice(pool) * random.uniform(0.99, 1.01)
        ny = random.choice(pool) * random.uniform(0.99, 1.01)
        nz = random.choice(pool) * random.uniform(0.99, 1.01)
        new_gen.append((nx, ny, nz))
    solutions = new_gen

best = ranked[0][1]
print(f"\nOptimized Values: x={best[0]}, y={best[1]}, z={best[2]}")
```

```python
import random

def f(x):
    return x**2 - 10*x + 25

def fitness(x):
    val = f(x)
    return 1 / (1 + val)

population = [random.uniform(-100, 100) for _ in range(200)]

for gen in range(500):
    ranked = sorted([(fitness(x), x) for x in population], key=lambda x: x[0], reverse=True)
    
    parents = [ranked[i][1] for i in range(50)]
    new_gen = []
    
    while len(new_gen) < 200:
        p1, p2 = random.choice(parents), random.choice(parents)
        child = (p1 + p2) / 2
        if random.random() < 0.1:
            child += random.uniform(-1, 1)
        new_gen.append(child)
    population = new_gen

best_x = sorted([(fitness(x), x) for x in population], key=lambda x: x[0], reverse=True)[0][1]
print(f"Best x: {best_x}, Min Value: {f(best_x)}")
```

```python
import random

items = [
    {'w': 2, 'v': 12}, # A
    {'w': 1, 'v': 10}, # B
    {'w': 3, 'v': 20}, # C
    {'w': 2, 'v': 15}  # D
]
CAPACITY = 5

def get_fitness(chromosome):
    weight = sum(chromosome[i] * items[i]['w'] for i in range(len(items)))
    value = sum(chromosome[i] * items[i]['v'] for i in range(len(items)))
    return value if weight <= CAPACITY else 0

population = [[random.randint(0, 1) for _ in range(4)] for _ in range(50)]

for gen in range(100):
    ranked = sorted(population, key=lambda c: get_fitness(c), reverse=True)
    parents = ranked[:20]
    
    new_gen = []
    while len(new_gen) < 50:
        p1, p2 = random.choice(parents), random.choice(parents)
        cp = random.randint(1, 3)
        child = p1[:cp] + p2[cp:]
        if random.random() < 0.1:
            mi = random.randint(0, 3)
            child[mi] = 1 - child[mi]
        new_gen.append(child)
    population = new_gen

best_chrom = sorted(population, key=lambda c: get_fitness(c), reverse=True)[0]
print(f"Best Chromosome: {best_chrom}, Max Value: {get_fitness(best_chrom)}")
```
