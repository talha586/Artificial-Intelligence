```python
#QUESTION 1
import heapq
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': ['G'],
    'E': ['G'],
    'F': ['G']
}

heuristic = {
    'A': 6, 'B': 4, 'C': 2, 'D': 3, 'E': 1, 'F': 5, 'G': 0
}

def greedy_best_first_search(graph, heuristic, start, goal):
    pq = [(heuristic[start], start, [start])]
    visited = set()
    expansion_order = []

    while pq:
        h_val, current, path = heapq.heappop(pq)
        
        if current in visited:
            continue
            
        expansion_order.append(current)
        visited.add(current)
        
        if current == goal:
            return expansion_order, path
        
        for neighbor in graph.get(current, []):
            if neighbor not in visited:
                heapq.heappush(pq, (heuristic[neighbor], neighbor, path + [neighbor]))
    
    return expansion_order, None

expansion, final_path = greedy_best_first_search(graph, heuristic, 'A', 'G')
print(f"Order of node expansion: {expansion}")
print(f"Path from start to goal: {final_path}")
```

```python
#QUESTION 2
import heapq

weighted_graph = 
{
    'A': [('B', 2), ('C', 4)],
    'B': [('D', 3), ('E', 1)],
    'C': [('F', 5)],
    'D': [('G', 2)],
    'E': [('G', 4)],
    'F': [('G', 1)]
}

heuristic = {
    'A': 6, 'B': 4, 'C': 2, 'D': 3, 'E': 1, 'F': 5, 'G': 0
}

def a_star_search(graph, heuristic, start, goal):
    pq = [(0 + heuristic[start], 0, start, [start])]
    visited = {}

    print(f"{'Node':   } {'g(n)':   } {'h(n)':   } {'f(n)':   }")
    
    while pq:
        f_val, g_val, current, path = heapq.heappop(pq)
        
        if current in visited and visited[current] <= g_val:
            continue
            
        visited[current] = g_val
        print(f"{current:   } {g_val:   } {heuristic[current]:   } {f_val:   }")
        
        if current == goal:
            return path, g_val
        
        for neighbor, cost in graph.get(current, []):
            new_g = g_val + cost
            new_f = new_g + heuristic[neighbor]
            heapq.heappush(pq, (new_f, new_g, neighbor, path + [neighbor]))

    return None, float(0)

opt_path, total_cost = a_star_search(weighted_graph, heuristic, 'A', 'G')
print(f"\nOptimal Path: {opt_path}")
print(f"Total Cost: {total_cost}")
```

```python
#QUESTION 3
class Node:
    def __init__(self, value=None):
        self.value = value
        self.children = []

def minimax(node, depth, isMaxPlayer):
    if not node.children:
        return node.value

    if isMaxPlayer:
        best_val = float('-inf')
        for child in node.children:
            value = minimax(child, depth + 1, False)
            best_val = max(best_val, value)
        node.value = best_val
        print(f"MAX node at depth {depth} value: {node.value}")
        return best_val
    else:
        best_val = float('inf')
        for child in node.children:
            value = minimax(child, depth + 1, True)
            best_val = min(best_val, value)
        node.value = best_val
        print(f"MIN node at depth {depth} value: {node.value}")
        return best_val

root = Node()
min_l, min_r = Node(), Node()
root.children = [min_l, min_r]
min_l.children = [Node(3), Node(5)]
min_r.children = [Node(2), Node(9)]

best_score = minimax(root, 0, True)
print(f"Best value for root: {best_score}")
print("Optimal move for MAX: Move to Left MIN child")
```

```python
#QUESTION 4
class Node:
    def __init__(self, value=None):
        self.value = value
        self.children = []

eval_count = 0
prune_count = 0

def alpha_beta(node, depth, alpha, beta, isMaxPlayer):
    
    if not node.children:
        eval_count += 1
        return node.value

    if isMaxPlayer:
        value = float('-inf')
        for child in node.children:
            value = max(value, alpha_beta(child, depth + 1, alpha, beta, False))
            alpha = max(alpha, value)
            if beta <= alpha:
                prune_count += 1
                print(f"Pruning at MAX node (depth {depth}): alpha={alpha}, beta={beta}")
                break
        return value
    else:
        value = float('inf')
        for child in node.children:
            value = min(value, alpha_beta(child, depth + 1, alpha, beta, True))
            beta = min(beta, value)
            if beta <= alpha:
                prune_count += 1
                print(f"Pruning at MIN node (depth {depth}): alpha={alpha}, beta={beta}")
                break
        return value

root = Node()
m1, m2 = Node(), Node()
root.children = [m1, m2]
m1.children = [Node(4), Node(3), Node(8)]
m2.children = [Node(5), Node(6), Node(7)]

res = alpha_beta(root, 0, float('-inf'), float('inf'), True)
print(f"Final Value: {res}")
print(f"Nodes evaluated: {eval_count}")
print(f"Pruning occurrences: {prune_count}")
```

```python
#QUESTION 5
graph = 
{
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['E'],
    'D': ['F'],
    'E': ['F'],
    'F': []
}

values = {
    'A': 10, 'B': 15, 'C': 13, 'D': 20, 'E': 18, 'F': 25
}

def hill_climbing(graph, values, start):
    current = start
    path = [current]
    
    while True:
        neighbors = graph.get(current, [])
        if not neighbors:
            break
            
        best_neighbor = max(neighbors, key=lambda x: values[x])
        
        if values[best_neighbor] > values[current]:
            print(f"Moving from {current}({values[current]}) to {best_neighbor}({values[best_neighbor]})")
            current = best_neighbor
            path.append(current)
        else:
            break
            
    return current, path

final_state, steps = hill_climbing(graph, values, 'A')
print(f"Steps: {steps}")
print(f"Final state: {final_state}")
print(f"Status: This is a Global Maximum as {values[final_state]} is the highest value in the set.")
```
