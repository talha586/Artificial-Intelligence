```python
#task3
def replace_first_char(s):
    if not s:
        return s
    first_char = s[0]
    lower = first_char.lower()
    upper = first_char.upper()
    temp = s.replace(lower, '@').replace(upper, '@')
    return first_char + temp[1:]

print(replace_first_char('Ooogle'))
```

```python
#task4
def jumble_strings(a, b):
    new1 = b[:2] + a[2:]
    new2 = a[:2] + b[2:]
    return new1 + ' ' + new2
print(jumble_strings('mix','pod'))
```

```python
#task5
def match_first_last(data):
    count = 0
    for x in data:
        if len(x) >= 2 and x[0] == x[-1]:
            count += 1
    return count
print(match_first_last(['aba','xyz']))
```

```python
#task6
def group_strings(string):
    temp_list = []
    other_list = []
    for s in string:
        if s.startswith('a'):
            temp_list.append(s)
        else:
            other_list.append(s)
    temp_list.sort()
    other_list.sort()
    return temp_list + other_list
print(group_strings(['ali','talha','abdullah']))
```

```python
#task7
def sort_by_last(tuples):
    
    def get_last(t):
        return t[-1]
        
    return sorted(tuples, key=get_last)
    
print(sort_by_last([(1,7),(1,3),(2,2)]))
```

```python
#task9
def flatten_list(lists):
    flat = []
    for sublist in lists:
        for item in sublist:
            flat.append(item)
    return flat
print(flatten_list([[1,2,3],[4,5]]))
```

```python
#task12
def binary_search(lst, target):
    low = 0
    high = len(lst) - 1
    while low <= high:
        mid = (low + high) // 2
        if lst[mid] == target:
            return mid
        elif lst[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
print(binary_search([1,2,3,4],4))
```

```python
#task13
def merge_sort(lst):
    if len(lst) <= 1:
        return lst
    mid = len(lst) // 2
    left = merge_sort(lst[:mid])
    right = merge_sort(lst[mid:])
    result = []
    i = 0
    j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result
print(merge_sort([4,2,3,5,1,9]))
```

```python
#task14
def quick_sort(lst):
    if len(lst) <= 1:
        return lst
    pivot = lst[len(lst) // 2]
    left = [x for x in lst if x < pivot]
    middle = [x for x in lst if x == pivot]
    right = [x for x in lst if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)
print(quick_sort([4,2,3,5,1,9]))
```

```python
#task15
def frequency_counter(lst):
    freq = {}
    for item in lst:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1
    return freq
print(frequency_counter([1,2,2,2,4,5]))
```

```python
#task19
def most_common_values(d):
    if not d:
        return []
    val_counts = {}
    for v in d.values():
        if v in val_counts:
            val_counts[v] += 1
        else:
            val_counts[v] = 1
    max_count = max(val_counts.values())
    most_common_vals = [k for k, v in val_counts.items() if v == max_count]
    return [k for k, v in d.items() if v in most_common_vals]

print(most_common_values())
```

```python
#task20
def combine_dicts(d1, d2):
    combined = d2.copy()
    combined.update(d1)
    return combined
print(combine_dicts({'a':1,'b':2},{'b':3,'c':4}))
```

```python
#task21
def filter_dicts(dict_list, target_key):
    return [d for d in dict_list if target_key in d]
print(filter_dicts([{'id':1,'name': 'ali'},{'id':2,'name':'talha'}],'id'))
```

```python
#task22
def find_duplicates(lst):
    temp = set()
    dupl = set()
    for item in lst:
        if item in temp:
            dupl.add(item)
        else:
            temp.add(item)
    return list(dupl)
print(find_duplicates([1,2,3,2,4,5,1]))
```
